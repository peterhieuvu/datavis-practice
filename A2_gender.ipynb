# Boilerplate

We first have to load in the modules we need. We need pandas and numpy for working with the data and altair for the visualization.

```python
import pandas as pd
import numpy as np
import altair as alt
```

# Recession Data
Here we create the data for the recession regions so that we can highlight the recession sections

```python
recessions = [
    {
            "start": 2001,
            "end": 2002,
            "event": "Early 2000s Recession"
    },
    {
            "start": 2007,
            "end": 2009,
            "event": "The Great Recession"
    }
]
df_rec = pd.DataFrame(recessions)

highlight = alt.Chart(df_rec).mark_rect(fillOpacity=.4).encode(
    x='start:O',
    x2='end:O',
    color=alt.Color('event:N', legend=alt.Legend(title='Recession'))
)
```

# Making some real charts

We first melt the data so that we can graph trademarks by origin

## Trademarks

```python
df_usa = df_usa.rename(columns={
    'Unemployment, female (% of female labor force) (modeled ILO estimate)': 'Female',
    'Unemployment, male (% of male labor force) (modeled ILO estimate)': 'Male',
    'Unemployment, youth female (% of female labor force ages 15-24) (modeled ILO estimate)': 'Youth female',
    'Unemployment, youth male (% of male labor force ages 15-24) (modeled ILO estimate)': 'Youth male'
})
col = list(df_usa.columns[3:])

df_pat = pd.melt(df_usa, id_vars=['Year'], value_vars=col[41:43],
                   var_name='Unemployment', value_name='Unemployment Rate')

# Remove 2017 year
df_pat = df_pat[(df_pat['Year'] != 2017) & (df_pat['Year'] > 1992)]


pat = alt.Chart(df_pat).mark_line().encode(
    x=alt.X('Year:O'),
    y='Unemployment Rate',
    color=alt.Color('Unemployment', scale=alt.Scale(domain=['Female', 'Male'],
                                                   range=['#f781bf', '#9acae9']),
                   legend=alt.Legend(title='Sex'))
    #tooltip=['Item1', 'Item2', 'Item3', 'Item4']
).properties(
    title='Unemployment rates by year (% of relevant labor force)',
    width=760,
    height=400,
)#.interactive()

(pat + highlight).resolve_scale(color='independent')
```

## Patents

```python
df_pat = pd.melt(df_usa, id_vars=['Year'], value_vars=col[43:45],
                   var_name='Unemployment', value_name='Unemployment Rate')

# Remove 2017 year
df_pat = df_pat[(df_pat['Year'] != 2017) & (df_pat['Year'] > 1992)]


pat = alt.Chart(df_pat).mark_line().encode(
    x='Year:O',
    y='Unemployment Rate',
    color=alt.Color('Unemployment', scale=alt.Scale(domain=['Youth female', 'Youth male'],
                                                   range=['#f781bf', '#9acae9']),
                   legend=alt.Legend(title='Sex'))
    #tooltip=['Item1', 'Item2', 'Item3', 'Item4']
).properties(
    title='Youth unemployment rates by year (% of relevant labor force, 15-25)',
    width=760,
    height=400,
)#.interactive()

(pat + highlight).resolve_scale(color='independent')
```

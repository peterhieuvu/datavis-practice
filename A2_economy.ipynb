```python
import pandas as pd
import numpy as np
import altair as alt
```

# Recession Data
Here we create the data for the recession regions so that we can highlight the recession sections

```python
recessions = [
    {
            "start": 1970,
            "end": 1971,
            "event": "Recesion of 1969-1970"
    },
    {
            "start": 1973,
            "end": 1975,
            "event": "Recession of 1973"
    },
    {
            "start": 1980,
            "end": 1981,
            "event": "1980 Recession"
    },
    {
            "start": 1990,
            "end": 1991,
            "event": "Early 1990s Recession"
    },
    {
            "start": 2001,
            "end": 2002,
            "event": "Early 2000s Recession"
    },
    {
            "start": 2007,
            "end": 2009,
            "event": "The Great Recession"
    }
]
df_rec = pd.DataFrame(recessions)

highlight = alt.Chart(df_rec).mark_rect(fillOpacity=.4).encode(
    x='start:O',
    x2='end:O',
    color=alt.Color('event:N', legend=alt.Legend(title='Recession'))
)
highlight
```

# Loading data and sanity check

We now load in the data and check the data

```python
df = pd.read_csv('economy-and-growth.csv')
df.head()
```

```python
df_usa = df[df['Country Code']=='USA'].sort_values(by=['Year'])
print(len(df_usa))
df_usa.astype(bool).sum(axis=0)
```

```python
#df_temp = df_usa.copy()
#df_temp['ptemp'] = df_temp['Patent applications, residents'].diff()

col = list(df_usa.columns[3:])

df_usa = df_usa[(df_usa['Year'] != 2017) & (df_usa['Year'] > 1964)]

#7, 8
#13, 14
#18
#21, 22
#36
cur = col[18]
print(cur)
temp = alt.Chart(df_usa).mark_line().encode(
    x='Year:O',
    y=cur,
    #color='Third Dim',
    #tooltip=['Item1', 'Item2', 'Item3', 'Item4']
).properties(
    title='Gross domestic savings by year (current US$)',
    width=760,
    height=400
)#.interactive()
(temp + highlight).resolve_scale(color='independent')
```

# Making some real charts

We first melt the data so that we can graph trademarks by origin

## Trademarks

```python
df_usa = df_usa.rename(columns={
    'Exports of goods and services (BoP, current US$)': 'Imports of goods and services',
    'Imports of goods and services (BoP, current US$)': 'Exports of goods and services'
})
df_melted = pd.melt(df_usa, id_vars=['Year'], value_vars=['Exports of goods and services',
                                                          'Imports of goods and services'],
                   var_name='Imports and Exports', value_name='Value (Current US$)')

# Remove 2017 year
df_melted = df_melted[df_melted['Year'] != 2017]
```

```python
tdmk = alt.Chart(df_melted).mark_line().encode(
    x='Year:O',
    y='Value (Current US$)',
    color=alt.Color('Imports and Exports', legend=alt.Legend(title='Imports and Exports'))
    #tooltip=['Item1', 'Item2', 'Item3', 'Item4']
).properties(
    title='Imports and exports of goods and service by year',
    width=760,
    height=400,
)#.interactive()

(tdmk + highlight).resolve_scale(color='independent')
```

## Patents

```python
df_pat = pd.melt(df_usa, id_vars=['Year'], value_vars=col[41:45],
                   var_name='Unemployment', value_name='Unemployment Rate')

# Remove 2017 year
df_pat = df_pat[(df_pat['Year'] != 2017) & (df_pat['Year'] > 1964)]


pat = alt.Chart(df_pat).mark_line().encode(
    x='Year:O',
    y='Unemployment Rate',
    color=alt.Color('Unemployment', legend=alt.Legend(title='Unemployment'))
    #tooltip=['Item1', 'Item2', 'Item3', 'Item4']
).properties(
    title='Title',
    width=760,
    height=400,
)#.interactive()

(pat + highlight).resolve_scale(color='independent')
```
